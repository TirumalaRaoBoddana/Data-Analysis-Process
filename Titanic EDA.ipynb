## **Correcting the data types of the columns or variables**

```python
titanic.dtypes
```

```python
#change the survived,pclass,sex and embarked to category
titanic[["Survived","Pclass","Sex","Embarked"]]=titanic[["Survived","Pclass","Sex","Embarked"]].astype("category")
titanic.dtypes
```

## **Checking and Handling missing values**

```python
for column in titanic.columns:
  print(column,"\t\t\t",titanic[column].isnull().sum())
```

- Age,Cabin and Embarked have missing values.
- we can replace the missing values of the age column with mean
- In case of the Cabin column it is not possible to fill the null values because it is not containing the numerical or categorical values and Cabin column has no relationship with the target variable Survived so we can drop the Cabin column
- Embarked column has 2 missing value we can fill them using mode because the column is categorical
- we can drop the passengerId.it is useful in predicting the target variable


```python
#replacing the nan values in age with mean age of the passengers
mean_age=titanic["Age"].mean()
titanic["Age"].fillna(mean_age,inplace=True)
```

```python
#checking for null values in Age colums
titanic["Age"].isnull().sum()
```

```python
#dropping the passengerId and the Cabin column from the dataset
titanic.drop(columns={"PassengerId","Cabin"},inplace=True)
titanic.columns
```

```python
#replacing the null values of the Embarked column with the mode of the Embarked column
embarked_mode=titanic["Embarked"].mode()[0]
titanic["Embarked"].fillna(embarked_mode,inplace=True)
```

```python
#checking for the null values
titanic.isnull().sum()
```

```python
#convert age column to int
titanic["Age"]=titanic["Age"].astype("int")
titanic.dtypes
```

## **Outlier Detection:**

```python
#detecting outliers in age column
sns.boxplot(titanic["Age"])
plt.title("outliers in Age column")
```

```python
#removing outliers in age column using
iqr=titanic["Age"].quantile(0.75)-titanic["Age"].quantile(0.25)
upper_limit=titanic["Age"].quantile(0.75)+(1.5*iqr)
lower_limit=titanic["Age"].quantile(0.25)-(1.5*iqr)
titanic=titanic[(titanic["Age"]>lower_limit) & (titanic["Age"]<upper_limit)]
titanic.reset_index(drop=True,inplace=True)
sns.boxplot(titanic["Age"])
```

```python
#detecting outliers in Fare column
sns.boxplot(titanic["Fare"])
plt.title("outliers in Fare column")
```

```python
#removing outliers in Fare column using
iqr=titanic["Fare"].quantile(0.75)-titanic["Fare"].quantile(0.25)
upper_limit=titanic["Fare"].quantile(0.75)+(1.5*iqr)
lower_limit=titanic["Fare"].quantile(0.25)-(1.5*iqr)
titanic=titanic[(titanic["Fare"]>lower_limit) & (titanic["Fare"]<upper_limit)&(titanic["Fare"]>0)]
titanic.reset_index(drop=True,inplace=True)
sns.boxplot(titanic["Fare"])
```

# **Feature Engineering**

```python
#converting the Sex column numerical column using one-hot encoding for male-0 female-1
titanic["Sex"]=titanic["Sex"].map({"male":0,"female":1})
titanic["Sex"]
```

```python
#converting the Embarked column into numerical
titanic["Embarked"]=titanic["Embarked"].map({"C":0,"Q":1,"S":2})
titanic["Embarked"]
```

```python
titanic.drop(columns={"Name","Ticket"},inplace=True)
titanic
```

```python
#normalizing the age column using the z score normalization
titanic["Age"]=(titanic["Age"]-titanic["Age"].mean())/(titanic["Age"].std())
titanic["Age"]
```

```python
#normalizing the fare column in titanic dataset
titanic["Fare"]=(titanic["Fare"]-titanic["Fare"].mean())/(titanic["Fare"].std())
titanic["Fare"]
```
